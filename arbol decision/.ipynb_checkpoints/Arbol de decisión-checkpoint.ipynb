Se calcula la entropía 

```python
def calculate_entropy(df_label):
    classes,class_counts = np.unique(df_label,return_counts = True)
    entropy_value = np.sum([(-class_counts[i]/np.sum(class_counts))*np.log2(class_counts[i]/np.sum(class_counts)) 
                        for i in range(len(classes))])
    return entropy_value
```

Definir la función de calcular la ganancia de información: esta función toma tres parámetros, a saber, conjunto de datos, característica y etiqueta.

```python
def calculate_information_gain(dataset,feature,label): 
    # Calculate the dataset entropy
    dataset_entropy = calculate_entropy(dataset[label])   
    values,feat_counts= np.unique(dataset[feature],return_counts=True)
    
    weighted_feature_entropy = np.sum([(feat_counts[i]/np.sum(feat_counts))*calculate_entropy(dataset.where(dataset[feature]
                              ==values[i]).dropna()[label]) for i in range(len(values))])    
    feature_info_gain = dataset_entropy - weighted_feature_entropy
    return feature_info_gain
```

Definir la función de creación de árbol de decisión, se divide el conjunto de entrenamiento en subconjuntos cada vez más pequeños.

```python
def create_decision_tree(dataset,df,features,label,parent):
    
    datum =np.unique(df[label],return_counts=True)
    unique_data=np.unique(dataset[label])
    
    if len(unique_data) <= 1:
        return unique_data[0]
    
    elif len(dataset)==0:
        return unique_data[np.argmax(datum[1])]
    
    elif len(features) == 0:
        return parent
    else:
        parent = unique_data[np.argmax(datum[1])]
        
        item_values = [calculate_information_gain(dataset,feature,label) for feature in features]
        
        optimum_feature_index = np.argmax(item_values)
        optimum_feature = features[optimum_feature_index]
        decision_tree={optimum_feature:{}}
        features = [i for i in features if i != optimum_feature]
        
        for value in np.unique(dataset[optimum_feature]):
            min_data = dataset.where(dataset[optimum_feature]==value).dropna()
            
            min_tree = create_decision_tree(min_data,df,features,label,parent)
            
            decision_tree[optimum_feature][value]=min_tree
            
        return (decision_tree)
```

Se define un metodo para predecir si es o no es, debe tener dos parametros:
    los datos de entrenamiento y el arbol de decision que anteriormente
    creamos.

```python
def predict_diabetes(test_data,decision_tree):
    for nodes in decision_tree.keys():
        value = test_data[nodes]
        decision_tree = decision_tree[nodes][value]
        prediction = 0
        if type(decision_tree) is dict:
            prediction = predict_diabetes(test_data,decision_tree)
        else:
            prediction = decision_tree
            break;
    return prediction
```

se eligen las columnas de la glucosa y presion arterial y se guardan en una variable.
se pone una etiqueta con el nombre de la columna que queremos predecir
y se indica tambien si tendra algun padre, el cual es ninguno.

```python
features = df.columns[:-1]
label = 'Diabetes'
parent=None
features
```

Se crea el arbol de decision con el metodo anteriormente creado, donde se le envia el dataset completo, columnas armaran el arbol y que sera el resultado.
Despues se muestra el arbol armado, el cual nos muestra que caminos hay, indicando primero el valor de la glucosa y luego el valor de la presion arterial para por fin mostrar el resultado.
En este caso, es binario ya que puede ser o no.

```python
decision_tree = create_decision_tree(df,df,features,label,parent)
decision_tree
```

Se crea una variable que guarde los datos de prueba de la glucosa y la presion arterial,estos se convierten en forma de serie y se envian junto con el arbol de decision anteriormente creado para predecir si puede ser diabetico o no.

```python
sample_data = {'Glucose':44,'BloodPressure':20}
test_data = pd.Series(sample_data)
test_data
prediction = predict_diabetes(test_data,decision_tree)
prediction
```

Con el resultado obtenido, solo se usa un condicional para obtener la respuesta en frase, conociendo que 0 es no probable y 1 es que es probable.

```python
if(prediction!=0):
    resultado = "Puede ser"
else:
    resultado = "No es probable"
resultado
```
